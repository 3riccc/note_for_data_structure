# 图的遍历-深度优先

- 基本思想：略

## 性能分析
- 需要借助一个递归工作栈，所以空间复杂度为O|V|
- 对每个节点查找邻接节点的过程
    - 所以如果邻接矩阵表示，查找每个节点消耗时间O|V|，总时间为O|V方|
    - 如果邻接表表示，查找所有节点时间O|E|，总时间O|V|+|E|

## 深度优先生成树

- 只有连通图才能产生深度优先生成树
- 否则会产生一个森林
- 和广度优先一样，邻接表存储的图，深度优先生成树是不唯一的

## 图的遍历算法和图的连通性

- 遍历算法可以判断图的连通性

<hr/>
- 对于无向图
    - 连通的无向图，一次遍历就能访问所有节点
    - 非连通的无向图，一次遍历只能访问这个连通分量的所有节点
    
<hr/>
- 对于有向图：如果从初始点到图中的任意节点都有路径，则能访问到
