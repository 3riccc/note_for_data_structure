# 栈

## 基础知识
- 定义：线性表，只能一端插入删除
- 栈顶：线性表允许插入删除的一端
- 空栈：不含元素的空表

## 基础操作

```c
// 进站
Push(&S,&x)
// 出战
Pop(&S,&x)
// 读栈顶元素，不出战
GetTop(&S)
```

## 栈的顺序存储
- 称为顺序栈
- 一组地址连续的存储单元
- 有一个指针指向栈顶

```c
// 栈的顺序存储的实现
#define MaxSize 50
typedef struct{
    ElemType data[MaxSize];
    int top;
} SqStack;
```

- 栈顶指针：S.top，初始时S.top = -1
- 进栈操作：栈顶指针+1，赋值栈顶元素
- 出栈操作：先取出栈顶元素，再栈顶指针-1
- 栈空条件：S.top == -1
- 栈满条件：S.top == MaxSize-1
- 栈长：S.top + 1

## 共享栈

- 为了有效利用存储空间
- 存取数据时间复杂度O(1)

- top0 == -1，0号栈为空
- top1 == MaxSize,1号栈为空
- 0号进栈：top0先+1    再赋值
- 1号进栈：top1先-1    再赋值
