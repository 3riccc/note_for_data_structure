# 二叉树的搜索和线索二叉树

## -二叉树的搜索

### --二叉树的遍历
- 按某个路径，每个节点都仅访问一次

<hr/>
#### 先序遍历
- 递归遍历
- 从根节点开始
- 先序遍历左子树
- 先序遍历右子树

```c
void PerOrder(BiTree T){
    if(T!=NULL){
        visit(T);
        PerOrder(T->lchild);
        PerOrder(T->rchild);
    }
}
```

<hr/>
#### 中序遍历
- 递归遍历
- 中序遍历左子树
- 访问根节点
- 中序遍历右子树

```c
void InOrder(BiTree T){
    if(T!=NULL){
        InOrder(T->lchild);
        visit(T);
        InOrder(T->rchild);
    }
}
```

<hr/>
#### 后序遍历
- 递归
- 后序遍历左子树
- 后序遍历右子树
- 访问根节点

```c
void PostOrder(BiTree T){
    if(T != NULL){
        PostOrder(T->lchild);
        PostOrder(T->rchild);
        visit(T)
    }
}
```

### --递归算法与非递归算法的转换

pass
- ——不能pass，这个转换会涉及到一个求路径的知识点
<hr/>
- 借助栈完成递归和非递归的转换
- 先扫描根节点的所有左节点并将它们一一进栈


```c
// 中序遍历非递归算法，用栈
void InOrder2(BiTree T){
    InitStack(S); // 初始化栈
    BiTree p = T;  //T是遍历指针
    // p不空或者栈不空时循环
    while(p||!IsEmpty(S)){
        if(p){  // 根指针进栈，遍历左子树
            Push(S,p);
            p = p->lchild;
        }else{
            // 根指针退栈，访问根节点
            Pop(S,p);
            visit(p);
            p = p->rchild;
        }
    }
    
}
```

### --层次遍历


- 需要借助队列
- 需要会默写这个遍历算法：如下

```c
// 层次遍历算法
void LevelOrder(BiTree T){
    // 初始化队列
    InitQuene(Q);
    // 初始化树的节点指针
    BiTree p;
    // 把根节点放入队列
    EnQuene(T);
    // 如果队列不空
    while(!IsEmpty(Q)){
        // 取出队列头
        DeQuene(Q,p);
        // 访问节点
        visit(p);
        // 有孩子就入队
        if(p->lchild != NULL){
            EnQuene(p->lchild);
        }
        if(p->rchild != NULL){
            EnQuene(p->rchild);
        }
    }
}
```

### --由遍历序列确定二叉树

- 先序和中序：可以确定
    - 先序中的第一个节点是根节点
    - 中序里根节点会将序列分成两个字序列
    - 在先序中找到这两个字序列
    - 字序列的第一个节点是子树的根节点
    - 重复……
- 中序和后序：可以确定
    - 后序中的最后一个节点是根节点
    - 和先序类似，重复
- 层序和中序：可以确定
    - 读者自己思考

<hr/>
- 先序和后序：无法确定

## -线索二叉树

- 二叉树的遍历，说白了就是对非线性结构数据线性化的过程
- 用二叉链表表示的二叉树中存在着大量的空指针，所以用来指向起前驱和后继
- 线索二叉树的目的：加快查找前驱和后继的速度
- N个节点的二叉树中，空指针数量为N+1
    - 推导过程：
    - 每个叶节点两个空指针
    - 度为1的节点有一个空指针
    - N0=N2+N1（叶子节点数量=度为2的节点数量+1）
    - 总空指针：N = N0+N1+N2
- 规定：
    - 没有左子树：lchild指向节点的前驱
    - 没有右子树：rchild指向节点的后继
    - 还需要增加ltag和rtag表示指向的是前驱还是孩子

```c
// 线索二叉树的存储结构
typedef struct ThreadNode{
    ElemType data;
    struct ThreadNode *lchild,*rchild;
    int ltag,rtag;
}ThreadNode, *ThreadNode;
```
