## 综合题

```c
// 1
define MaxSize = 50;

//定义类型
typedef struct{
    int data[MaxSize];
    struct SqQuene *front,*rear;
    int tag;//0空，1满，2不空不满
} SqQuene;

// 入队
bool EnQuene(SqQuene &Q,int x){
    // 队满则不能入队
    if(Q.tag == 1) return false;
    // 入队
    Q.data[Q.rear] = x;
    Q.rear = (Q.rear + 1) % MaxSize;
    // 判断队满
    if(Q.rear == Q.front){
        Q.tag = 1;
    }else{
        Q.tag = 2;
    }
    return true;
}

// 出队
bool DeQuene(SqQuene &Q,int &x){
    // 队空不能出队
    if(Q.tag == 0) return false;
    // 出队
    x = Q.front;
    Q.front = (Q.front + 1)%MaxSize;
    // 判断队空
    if(Q.front == Q.rear){
        Q.tag = 0;
    }else{
        Q.tag = 2;
    }
    return true;
}

// 我不得不承认书上的思路比我的思路更好
// 但是懒得抄一遍了
// 直接截图在下面吧
```
