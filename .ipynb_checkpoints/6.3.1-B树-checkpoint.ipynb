# B树
- 要求掌握
- 基本特点
- 建立、插入、删除操作（不清楚这个是不是要求会写算法，还是懂这些操作的过程）

- 又称为多路平衡查找树
- B树的阶：所有孩子节点数的最大值
- m阶B树
    - 可以是空树
- m阶非空B树性质
    - 树中节点至多有m个子树
    - 根节点不是叶子节点，那就至少有两个子树
    - 除根节点和叶子节点外，所有节点至少有[m/2]颗子树
    - 

- 每个节点的孩子可以>=2个
- 每个节点可以存储多个元素
- 元素之间存在着排序关系（从大大小或者从小到大）
- 每个节点有两个或者三个孩子，称之为23树
- 每个节点要么没有孩子，要么有两三个孩子，不能只有一个孩子
