# 图的基本概念


## 1.定义

- 顶点和连边构成的集合
- |V|表示图的顶点个数，也叫图的阶
- |E|表示图的连边个数
- 线性表可以是空表，树可以是空树，但图不能是空图
    - 具体来说，图的顶点集一定非空
    - 但图的连边集可以为空

### 无向图PASS

### 简单图
- 不存在重复连边
- 不存在顶点到自身的连边
- 则称为简单图（数据结构中只讨论简单图）

### 多重图
- 定义和简单图相对

### 完全图
- 任意两个顶点之间都存在连边
- 有n个顶点图
    - 无向图有n(n-1)/2个连边
    - 有向图有n(n-1)个连边

### 子图
- 设有两个图G=(V,E)和G1 = (V1,E1)
- 其中V1是V的子集，G1是G的子集
- 称G1是G的子图
- **并非任何G的子集都是他的子图，因为子集可能不是图，这不是说子图不联通，而是E1中的节点可能不在V1中存在**

### 连通相关概念

#### 连通图
- 无向图中
- 任意两个顶点之间都有一条边链接的图

#### 连通分量
- 把整个图看作串了珠子的绳子
- 把任意顶点提起，连通图是一个整体。
- 非连通图散成若干个小的整体
- 每一个小的整体就是整幅图的一个连通分量

#### 极小连通子图
- 子图
- 要保持图连通
- 也要让边数最少的子图


#### 强连通图
- 有向图中
- 任意节点v到任意节点w有路径
- w到v也有路径
- 就是强连通图

#### 强连通分量
- 有向图中的极大强连通子图

### 顶点的度、入度、出度
- 定义pass
- 无向图的所有度加和=2e（e是连边数量）
- 有向图入度出度定义pass
- 有向图中所有入度出度之和为e（e是边数）
    - 这是因为每条边都有一个起点和终点

### 边的权和网（不重要）
- 边上可以带权
- 边上带权的图叫做网

### 稠密图和稀疏图（不重要）
- 边很少的图称作稀疏图
- 有多少才算呢？
    - 一般来说，判断标准：|E| < |V| * log|V| 成立，就说是稀疏的

### 路径、路径长度、环
- 路径是经过的指顶点序列
- 路径长度是经过的连边条数（带权的叫做带权路径长度）
- 如果n个节点，边数 >= n，一定有环

### 简单路径、简单回路（不重要）
- 顶点不重复出现的路径叫简单路径
- 回路中，除了第一个顶点出现两次，其他顶点都只出现一次的路径叫做简单回路

### 距离
- 两点之间的最短路径长度
- 如果两点之间不存在路径，距离计作正无穷


### 有向树
- 首先是个有向图
- 只有一个顶点的入度为0
- 其余所有顶点入度为1
