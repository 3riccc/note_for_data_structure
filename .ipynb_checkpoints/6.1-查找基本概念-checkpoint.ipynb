### 查找

在数据集中找到符合某种条件的数据元素

找到了：查找成功

没找到：查找失败

# 平均查找长度
- 一次查找长度指的是：一次查找需要比较的关键字次数
- 平均查找长度：所有查找过程中，关键字比较次数的均值
- 记为ASL


### 查找表（查找结构）
- 用于查找的数据集合
- 由同一类型的元素构成
- 常用以下操作：
    - 查询某个元素是否在查找表中
    - 查找满足条件的元素的各种属性（看起来感觉不重要）
    - 在里面插入一个元素
    - 在里面删除一个元素

### 静态查找表
- 不涉及插入和删除的查找表
- 适合静态查找表的查找方法：
    - 折半
    - 顺序
    - 散列

### 动态查找表
- 涉及插入删除的查找表
- 适合动态查找表的查找方法：
    - 二叉排序树
    - 散列（散列全都适合啊，牛逼）
