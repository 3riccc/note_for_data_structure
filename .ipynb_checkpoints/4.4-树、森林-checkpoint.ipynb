# 树、森林

## 1.树

### 1.1三种常用的存储结构

<hr/>
#### 1.1.1双亲表示法

- 顺序存储、连续空间
- 根节点，下标为0，指针域为-1
- 利用了每个节点双亲唯一的性质，可以快速找到节点的双亲
- 但是找节点孩子的时候，需要遍历整个树

```c
// 双亲表示法的存储结构
define MAX_SIZE 100;

// 每一个节点的存储结构
typedef struct{
    ElemType data;
    int parent;
} PTNode;

// 树的存储结构
typedef struct{
    ElemType PTNode[MAX_SIZE];
    int n; //树中的节点数量
} PTree;
```

<hr/>
#### 1.1.2孩子表示法

- 将每个节点的孩子都用单链表链接起来形成线性结构
- N个节点就有N个孩子链表（叶节点孩子链表为空）
- 找孩子方便
- 找双亲不方便，需要遍历整个表

<hr/>
#### 1.1.3孩子兄弟表示法（二叉树表示法）
（我觉得孩子兄弟，这个名字更贴切，一个指针指孩子，一个指针指兄弟）

- 用二叉链表作为树的存储结构
- 每个节点包括：
    - 节点值
    - 指向他的第一个孩子的指针
    - 指向他兄弟的指针
- 优点：
    - 方便的实现树转化为二叉树
    - 容易查找节点的孩子节点
- 缺点：
    - 查找双亲麻烦

```c
// 孩子兄弟表示法（二叉树表示法）的存储结构
typedef struct{
    ElemType data;
    struct CSNode *firstchild,*nextsubling;
}CSNode, *CSNode;
```

## 2.树、森林与二叉树的转换

- 树->二叉树  的规则
    - 就是上面提到的孩子兄弟表示法
    - 左指针指孩子，右指针指兄弟
    - 根节点没有兄弟，所以转化后二叉树根节点没有右子树
    


- 森林->二叉树   的规则
    - 将森林中的所有树看作兄弟
    - 森林中第二棵树，填充第一棵树转化出来的右子树
    - 第三棵树填充第二棵的右子树

- 二叉树->森林  的规则

## 3.树和森林的遍历

- 都可以用对应二叉树的遍历来实现：直接把树和森林转化为二叉树，然后对二叉树用先序后序之类的算法




<hr/>
#### 树的遍历
- 先根遍历
    - 先看根节点
    - 再从左到右看每一个子树
    - 和二叉树的先根遍历几乎一样，同样去理解就好
- 后根遍历
    - 先从左到右看每一个子树
    - 在看根节点
- 层次遍历
    - 和二叉树思想相同

<hr/>
#### 森林的遍历
- 先序遍历
    - 看森林中第一棵树的根节点
    - （这里开始涉及到递归）先序遍历第一棵树的根节点的**子树森林**
    - （这里还在递归）先序遍历去掉第一棵树后，其他树构成的森林
- 中序遍历
    - （一开始就递归）中序遍历第一棵树根节点的子树森林
    - 访问第一棵树的根节点
    - （还在递归）中序遍历去掉第一棵树后的其他树构成的森林
