# 4.1树的基本概念

## ----基本定义

- N(>=0)个节点的有限集合
- 0个节点的叫做空树

<hr/>
**在非空的树中**
- 有且仅有一个根节点
- N>1时，其余节点可以看作m个子树，互不相交
- 是一种递归定义的逻辑结构

<hr/>
**树的特点：**
- 每个节点有且只有一个前驱节点，根节点除外
- 每个节点可以有0或多个后继节点
- 任何树的节点数比连边数多1

## ----基本术语

- 祖先节点：A到K路径上的任意节点，都是K的祖先节点
- 子孙节点：你是我祖先，我就是你子孙
- 双亲节点：比我高一层的祖先，直接的爸爸
- 孩子节点：你是我双亲（双亲等于爸爸），我就是你孩子
- 兄弟节点：和我有共同爸爸的节点是我的兄弟节点

<hr/>
- 节点的度：一个节点的孩子的数量
- 树的度：树中节点的最大度

<hr/>
- 分支节点：度大于零的节点，有孩子的节点，也叫非终端节点
- 叶子节点：度=0，没有孩子，也叫终端节点

<hr/>
- 节点的层次：根节点第一层，以此类推
- 节点的深度：就是层数
- 节点的高度：从叶子节点开始累加，与深度刚好相反

<hr/>
- 无序树：叶子节点的大小排列没有关系
- 有序树：任何叶子节点必须从左到右有次序，比如图4-1，如果交换了任何两个节点的位置，都成了一棵不同的树

<hr/>
- 路径：
    - 树中的路径必须是**从上往下**的
    - 由两个节点之间所经过的节点序列构成
- 节点路径长度：
    - 路径上所经过的**边的个树**
    - A到K的路径长度是3
- 树的路径长度：所有节点路径长度的总和

<hr/>
- 森林：m（>=0）棵互不相交的树的集合

## ----树的性质

- 节点数等于所有节点的度数+1
- 度为m的树，第i层，至多有m^(i-1)个节点
    - 其原因是如果没层都是m个子树，每个子树都有m个分支，那么到第i层节点树最大，就是m^(i-1)
    - 例如二叉树度为2，第一层1个，第二层两个
