- 结合顺序存储与链式存储

## 请注意
- 无向图需要的存储空间是O(|V|+2|E|)——这是因为每条边都出现了两次
- 有限图需要的存储空间是O(|V|+|E|)
<hr/>
- 找到他的所有连边
    - 邻接表可以迅速找齐
    - 在邻接矩阵中，需要扫描一行，花费O(n)时间复杂度
- 确定两个节点之间是否存在连边
    - 邻接矩阵可以迅速查找到
    - 邻接表需要在相应节点对应的边表中查找另一个节点
- 有向图中，求一节节点的出度
    - 计算对应邻接表中的节点个数即可
- 有向图，求一个节点的入度
    - 遍历全部邻接表
    - 或者采用逆邻接表的方式存储
- 一个图的邻接表不唯一，因为在创建邻接表的时候，一个节点的连边次数是无序的
