# 邻接多重表

- **无向图**的另一种链式存储结构
- 每条边用一个节点表示，里面含有这些信息
    - mark：标志域，可以用来标识这个连边是否被搜索过
    - ivex、jvex：依附这条边的两个顶点
    - ilink：下一条依附于顶点ivex的边
    - jlink：吓一条依附于顶点jvex的边
    - info：其他信息指针，不重要
- 没个顶点也用一个节点表示，它包含
    - data：顶点信息
    - firstedge：第一条依附于这个顶点的边
