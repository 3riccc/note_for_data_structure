## 简单选择排序
- 简明视频（2分钟）https://www.bilibili.com/video/av18176082?from=search&seid=1335076102506380621

## 方法

- 从第二个数到最后一个数，找到最小的数（找的方法是变量初始化为第一个数，然后依次比较）
- 和第一个数交换
- 从第三个数到最后一个数，找到最小的数
- 和第二个数交换
- 重复以上过程

## 分析

- 空间效率：常数个辅助单元，所以空间复杂度为O(1)
- 时间效率：
    - 元素移动的次数很少，不超过3(n-1)次，这是因为每次交换两个位置元素需要操作三次
    - 元素比较次数恒定，是n(n-1)/2次
    - 所以时间复杂度O(n方)    
- 稳定性：不稳定

## 大顶堆的插入

- 调整整个树，重新满足大顶堆的性质

## 分析
- 空间效率O（1）

    - 建堆时间O（n）
