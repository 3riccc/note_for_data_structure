# 顺序查找
- 又叫“线性查找”
- 一种最直观的查找方法

<hr/>
### 基本思想

- 从线性表的一边开始
- 逐个检查关键字是否满足条件
- 成功就返回元素在线性表中的位置
- 到末尾还没找到就返回失败

### 优缺点
- 缺点
    - 当n比较大的时候，平均查找长度大，效率低
- 优点
    - 对数据存储没有要求，顺序存储和链式存储都可以
    - 对有序性没有要求，有虚表和无虚表皆可
    - 对现行**链表**只能进行顺序查找
