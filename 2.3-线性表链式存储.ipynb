### 单链表的单个节点

```c
#include <stdio.h>

// typedef：给类型起别名

// 单链表元素结构题
typeof struct LNode{
    // 数据
    ElemType data;
    // 指向下一个节点的指针
    struct LNode *next;
};
```

## 头插法建立链表

```c
// 插头法建立链表
LinkList CreateList1(LinkList &L){
	// 从表尾到表头逆向建立单链表
	LNode *s,int x;
	// 一块内存空间，linklist类型的
	// 创建头节点
	L = (LinkList)malloc(sizeof(LNode));
	L->next = NULL;

	// 输出节点的值
	scanf("%d",&x);
	// 创建新节点
	while(x!=9999){
		s = (LNode*)malloc(sizeof(LNode));
		s->data = x;
		s->next = L->next;
		// L一直是头，头插法实际上是插在了第二位
		L->next = s;
		scanf("%d",&x);
	}
	return L;
}
```

- 好处：简单
- 坏处：生成顺序和存储顺序不一样

## 尾插法建立链表

```c
LinkList CreateList2(LinkList &L){
	// 设置元素类型为整形
    int x;
    // 建立头节点
    L = (LinkList)malloc(sizeof(LNode));
    // r是表尾指针
    // s是每次生成的新节点
    LNode *s,*r=L;
    // 输入节点的值
    scanf("%d",&x);
    // 生成链表
    while(x != 9999){
    	s = (LNode *)malloc(sizeof(LNode));
    	s->data = x;
    	// 上一个节点和新节点建立连接
    	r->next = s;
    	r = s;
    	scanf('%d',&x);
    }
    // 最后一个指针指向NULL
    r->next = NULL;
    return L
}
```

- 时间复杂度和头插法相同

# 静态链表

### 静态链表的基本结构



#### 用左边的线性表表示了右边的链表

```c
// 静态链表的结构,首先定义长度
# define MaxSize = 50
typedef struct{
    ElemType data;
    int next;
} SLinkList[MaxSize]
// Slinklist 前面的s值得是static的意思，我猜
```

- 以next = -1 作为结束标志
- 插入/删除操作与动态链表相同
    - 只需要修改指针，**不需要移动元素**
- 没有单链表使用方便
- 一般在一些不支持指针的语言中才会用（如basic）
- 长度一开始就固定死了
