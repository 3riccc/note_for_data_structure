# 图的存储方法：邻接矩阵

**核心思路**
- 用一个一维数组存储节点信息
- 用一个二维数组存储连边信息（也就是邻接矩阵）

## 请注意

- 邻接矩阵存储的空间复杂度是O(n方)，其中n是顶点数
- 无向图的邻接矩阵是一个对称阵，所以实际存储的时候只需要存储上三角元素就行
- 无向图的第i行非零元素个数就是第i个节点的度
- 有向图
    - 第i行非零元素的个数是节点的出度
    - 第i列非零元素的个数是节点的入度
- 优点和局限
    - 优点：容易确定两个节点之间是否存在链接
    - 局限：要确定一共有多少边，就需要按行检索，非常费劲
- 稠密图适合用邻接矩阵表示
