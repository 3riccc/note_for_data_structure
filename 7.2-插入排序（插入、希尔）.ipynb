# 插入排序

<hr/>
## 直接插入排序

- 因为是这一系列插入算法中最简单的，所以也叫简单插入排序


### 基本操作

把一个记录插入到一个已经排好序的有虚表中

得到一个新的有序表（记录数增加1）

### 算法分析

- 空间复杂度：使用了常数个缓冲区，空间复杂度O(1)

- 时间复杂度：向有序子表中逐个插入的操作进行了n-1次，每次都分成比较和移动两个部分，所以
    - 最好的情况下，表中元素已经有序，只需要和每个比较一次，不用移动，时间复杂度O(n)
    - 最坏的情况下，表中元素逆序，总比较次次数达到最大
    
    - 平均情况下，表中元素随机，总的比较次数与移动次数约为 n方/4
    - 所以时间复杂度为O(n方)

### 稳定性
- 相同元素相对位置不变化，稳定

### 适应性
- 顺序存储和链式存储
- 大部分排序算法都只适用于顺序存储的线性表

### 适应范围
- 基本有序
- 数据量不大

## 折半插入排序

- 直接插入排序总结
    - 进行了两项工作
    - 从前面的子表中找出合适的位置
    - 给位置腾出空间（后移）在插入
    - 一边比较一边移动


- 折半插入的改进
    - 先找出待插入的位置（折半查找，这是最重要的区别）
    - 统一的移动插入位置之后的所有元素（这段有点混乱，啥意思）

## 希尔排序


### 基本思想
- 先把排序表分成若干个小区间[i,i+d,i+2d,...,i+kd]
- 分别进行直接插入排序
- 整个表基本有序后
- 再整个表进行一次直接插入排序


### 过程
- 取一个小于n的步长d1 = n/2
- 所有距离为d1倍数的，放在一个组里
- 在各组中进行直接插入排序
- 再取第二个步长d2 = d1/2
- 直到去到dn=d(n-1)/2 = 1

### 分析

- 时间复杂度
    - 最坏的情况下，O(n方)
    - 好的情况，没有一般结论
- 空间复杂度
    - O(1) ：常数个辅助单元
- 稳定性
    - 不稳定
- 适用性
    - 仅适用于顺序存储，不适用于链式存储
