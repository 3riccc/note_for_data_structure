# 哈夫曼树

<hr/>
## 定义

- 带权路径长度
    - 叶节点有一个表示权重的值
    - 路径长度：从根节点走到叶节点的边数
    - 带权路径长度：路径长度 * 节点的权
- 哈夫曼树
    - 含有N个带权叶节点的二叉树
    - 带权路径最小的树成为哈夫曼树
    - 也称作最优二叉树


<hr/>
## 构造

- 给定N个权重为W1，W2...WN的节点
- 把这N个节点看作N棵只有一个节点的树
- 循环：
    - 从所有树中，找到两个权重最小的
    - 构造一个新节点，让这两个节点作为新节点的左右子树
    - 新节点的权重是这两个节点的权重的和
    - 用新树替换构造新树用到的两个树

## 哈夫曼编码

- 前缀编码：
    - 没有一个编码是另一个编码的前缀。
    - 这个意思是说，不会出现一个编码的前面三个字节是另一个编码
    - 比如0，10，01，110是前缀编码，但11就不能融入进去，因为11是110的前缀
    - 哈夫曼编码就是典型的前缀编码


#### 从哈夫曼树到哈夫曼编码
- 每一个字符是一个节点
- 权重是他出现的频次
- 构造出对应的哈夫曼树
- 编码就是从根节点到叶节点的路径标记序列：
    - 0代表转向左孩子
    - 1代表转向右孩子

这颗哈夫曼树的带权路径长度

WPL = 45*1 + （12+13+16）* 3 + （5+9） * 4 = 224
