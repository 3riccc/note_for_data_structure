# 二叉树的概念

## ----定义和主要特性

- 每个节点至多两个子树
- 是有序树，子树有左右之分，不能颠倒

<hr/>
**二叉树与度为2的树的区别**
- 度为2，树至少有3个节点，而二叉树可以为空
- 度为2的树左右节点是相对而言的，如果只有一个孩子节点，那就不分左右，而二叉树无论有几个孩子节点，均分左右

## ----几个特殊的二叉树


<hr/>
**满二叉树**
- 高度为h
- 含有2^h - 1个节点
- 每一行都是满的
- 对于编号为i的节点，如果有双亲，双亲为i/2，如果有左孩子，左孩子2i，右孩子2i+1

<hr/>
**完全二叉树**
- 如果节点编号n<=i/2，他就是分支节点，否则就是叶子节点
- 满二叉树的最后一层没满
- 叶子节点只可能在层次最大的两层出现
- 一旦编号为i的节点只有左孩子，那么i之后的节点全都是叶子节点
- 只可能有一个度为1的节点，且这个节点只有左孩子
- 如果n为奇数，所有节点都有左右孩子
- 如果n为偶数，n/2号节点只有左孩子
- 度为1的节点只有左孩子

<hr/>
**二叉排序树**
- 可能是空二叉树
- 否则左子树上所有节点的关键字 < 根节点关键字
- 右子树所有节点的关键字 > 根节点关键字

<hr/>
**平衡二叉树**
- 任意节点的左右子树深度之差不超过1

## ----二叉树的性质

- **非空二叉树上的叶子节点数 = 度为2的节点数 + 1**
- 第K层至多有2^(k-1)个节点
- 高度为H的二叉树至多有2^H-1 个节点
- 对完全二叉树编号，则
    - 节点i所在层次深度为log2i + 1
    - 具有N个节点的完全二叉树，高度为[log2(N+1)]，或者[log2N] + 1    []是取整的意思，向上取整？？？

## ----二叉树的存储结构

### 1.顺序存储
- 用一组地址连续的存储单元从上到下从左到右存储**完全二叉树**上的节点元素，也就是将完全二叉树的编号为i的节点，存储在下标为i-1的分量中
- 根据二叉树的性质，完全二叉树和满二叉树都适合顺序存储，因为序号关系可以完全表示节点之间的逻辑关系
