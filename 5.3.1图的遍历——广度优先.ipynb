# 图的广度优先遍历

- 不是递归算法
- 分层
- 借助队列
- 类似于树的层次遍历

参考资料，小甲鱼教程

```c
// 代码实现
// 保存被访问过的数组代号
bool visited[MAX_NUM];

// 遍历前的准备函数
bool BFSTraverse(Graph G){
    // 记录数组的初始化
    for(int i =0 ;i <G.vexnum;i++){
        visited[i] = FALSE;
    }
    // 初始化辅助队列
    InitQuene(Q);
    // 从零号开始遍历
    for(int i =0 ;i <G.vexnum;i++){
        BFS(G,i)
    }
}

// 真正的遍历控制函数
void BFS(Graph G,int v){
    // 访问这个节点
    visit(v); // 这里很神奇，visit接受的参数竟然是一个整数，而不是一个节点
    // 记录
    visited[v] = TRUE;
    // 顶点入队
    EnQuene(Q,v);
    // 如果队列不为空
    while(!IsEmpty(Q)){
        DeQuene(Q,v);// 出队？？？？？为什么刚入队就出队了？——你在疑惑什么，队列就是先入先出啊，你这都不记得？？？
        // 检查所有邻接节点
        for(w = FirstNeighbor(G,v);w>=0;w=NextNeighbor(G,v,w)){
            // 如果这个节点被没被访问过
            if(!visited[w]){
                visit(w)
                visited[w] = TRUE;
                // 访问完了入队
                EnQuene(Q,w)
            }
        }
    }
}
// got it
```

## 广度优先性能分析
### 空间复杂度
- 需要借助一个队列
- 在最坏的情况下，所有节点都要入队一次，空间复杂度O|V|

### 时间复杂度
- 采用邻接表存储时
    - 每个顶点入队一次
    - 时间复杂度O(|V|)
    - 搜索一个节点的邻接节点时候，每条边访问一次
    - 时间复杂度O(|E|)
    - 总的时间复杂度O(|V|+|E|)
- 采用邻接矩阵存储
    - 每个节点邻接节点的查询需要O|V|
    - 总时间复杂度O|V方|

## 广度优先算法求解最短路径问题
- 没时间做，我浪费太多时间了，下次继续

```c
// 这里为什么要用整型变量表示节点呢，因为用到了邻接表存储图，所以用整型变量表示节点索引
void BSF_mindistance(Graph G,int u){
    // d[i]表示从i节点到u节点的最短路径
    for(int i=0;i<G.vexnum;i++){
        d[i] = 正无穷;// 全部初始化为正无穷
    }
    d[u] = 0;
    // u节点入队
    EnQuene(Q,u);
    // 队列非空时循环
    while(!IsEmpty(Q)){
        DeQuene(Q,u);// u出队
        // 访问u节点的所有一阶邻居
        for(w=FirstNeighbor(G,u);w>=0;w=NextNeighbor(G,u,w)){
            // 如果这个邻居没有被访问到
            if(!visited[w]){
                visited[w] = TRUE;
                d[w] = d[u]+1;
                EnQuene(Q,w);
            }
        }
    }
}
```

## 广度优先生成树

- 邻接矩阵是唯一的，所以得到的树是唯一的
- 邻接表表示不唯一，所以得到的树不唯一
